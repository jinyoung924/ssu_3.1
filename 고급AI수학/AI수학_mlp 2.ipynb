# **문제 1: 이차원 다중 분류기**


## **[단계 1] 이차원 다중 분류기(2D Multi-class Classifier)을 위한 Data 생성 및 산점도 그리기**


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate data
N = 100
np.random.seed(78)
x1, x2 = 2 * np.random.rand(N) - 1, 2 * np.random.rand(N) - 1
x = np.row_stack((x1, x2))

tf1 = x1**2 + x2**2 < 0.7
tf2 = x1 + x2 < 0
tf3 = x1 - x2 < 0

t1 = np.logical_and(np.logical_and(tf1, tf3), tf2)
t2 = np.logical_and(np.logical_and(tf1, tf3), np.logical_not(tf2))
t3 = np.logical_and(np.logical_and(tf1, np.logical_not(tf3)), tf2)
t4 = np.logical_and(np.logical_and(tf1, np.logical_not(tf3)), np.logical_not(tf2))
t5 = np.logical_and(np.logical_not(tf1), tf2)
t6 = np.logical_and(np.logical_not(tf1), np.logical_not(tf2))

t = np.float64(np.row_stack((t1, t2, t3, t4, t5, t6)))

# Initial scatter plot
fig = plt.figure()
ax = plt.axes()
labels = t[1,:] + 2*t[2,:] + 3*t[3,:] + 4*t[4,:] + 5*t[5,:]
ax.scatter(x[0,:], x[1,:], c=labels, cmap='rainbow')
ax.set_aspect('equal')
plt.show()
```

## **[단계 2] 복층 신경회로망 구조 정의 및 경사하강법을 이용한 분류기 훈련**


### **신경회로망 구조  및 학습 형식 설정**
### (아래의 설정 예를 참고하여 분류기를 위해 설계한 모델을 작성합니다.)

### (설정 예)
#### - 입력, 은닉층, 출력층 노드 갯수: (D, L, M) = (784, 30, 10)
#### - 활성화 함수(Activation Functions)
* 은닉층(Hidden Layer): ReLU 함수
* 출력층(Output Layer): Softmax 함수

#### - 목적함수(비용함수): 최소제곱오차(MSE)
#### - 학습률(Learning Rate): 1.0
#### - 최대 반복수 (Max Iterations): 10,000
#### - 편미분(그래디언트) 계산 형식: bWx 형식의 행렬 표현법

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# 데이터 로드
mnist = fetch_openml('mnist_784', as_frame=False, parser='auto')
xx = mnist.data / 255.0  # 정규화, shape: (70000, 784)
labels = mnist.target.astype(int)  # Shape: (70000,)
N, D, L, M = len(labels), xx.shape[1], 30, len(np.unique(labels))  # N=70000, D=784, L=30, M=10

# 학습 및 테스트 데이터 분할
train_size = 60000
X_train, y_train = xx[:train_size], labels[:train_size]
X_test, y_test = xx[train_size:], labels[train_size:]
Y_train = np.eye(M)[y_train, :]  # 학습 레이블 원-핫 인코딩, shape: (60000, 10)
Y_test = np.eye(M)[y_test, :]    # 테스트 레이블 원-핫 인코딩, shape: (10000, 10)

# MLP 구현
class MLP:
    def __init__(self, input_size=784, hidden_size=30, output_size=10):
        # 가중치와 편향을 [-1, 1] 균등분포로 초기화
        self.W1 = np.random.uniform(-1, 1, (input_size, hidden_size))  # Shape: (784, 30)
        self.b1 = np.random.uniform(-1, 1, (1, hidden_size))          # Shape: (1, 30)
        self.W2 = np.random.uniform(-1, 1, (hidden_size, output_size))  # Shape: (30, 10)
        self.b2 = np.random.uniform(-1, 1, (1, output_size))          # Shape: (1, 10)

    def relu(self, x):
        """ReLU 활성화 함수: x > 0이면 x, 아니면 0"""
        return np.maximum(0, x)

    def relu_derivative(self, x):
        """ReLU의 도함수: x > 0이면 1, 아니면 0"""
        return np.where(x > 0, 1.0, 0.0)

    def softmax(self, x):
        """소프트맥스 함수: 클래스 확률로 변환"""
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward(self, X):
        """순전파: bWx 형식 (bias + W @ X)"""
        self.z1 = np.dot(X, self.W1) + self.b1  # Shape: (N, 30)
        self.a1 = self.relu(self.z1)  # Shape: (N, 30)
        self.z2 = np.dot(self.a1, self.W2) + self.b2  # Shape: (N, 10)
        self.a2 = self.softmax(self.z2)  # Shape: (N, 10)
        return self.a2

    def compute_loss(self, Y_pred, Y_true):
        """크로스엔트로피 손실 계산"""
        m = Y_true.shape[0]
        return -np.sum(Y_true * np.log(np.clip(Y_pred, 1e-10, 1.0))) / m

    def compute_accuracy(self, Y_pred, y_true):
        """정확도 계산: 올바른 예측의 백분율"""
        predictions = np.argmax(Y_pred, axis=1)
        return np.mean(predictions == y_true) * 100

    def backward(self, X, Y_true, learning_rate=1.0):
        """역전파: 그래디언트 계산 및 가중치 업데이트"""
        m = X.shape[0]
        # 출력층 그래디언트 (크로스엔트로피와 소프트맥스)
        d_z2 = self.a2 - Y_true  # Shape: (N, 10)
        d_W2 = np.dot(self.a1.T, d_z2) / m  # Shape: (30, 10)
        d_b2 = np.sum(d_z2, axis=0, keepdims=True) / m  # Shape: (1, 10)
        # 은닉층 그래디언트
        d_a1 = np.dot(d_z2, self.W2.T)  # Shape: (N, 30)
        d_z1 = d_a1 * self.relu_derivative(self.z1)  # Shape: (N, 30)
        d_W1 = np.dot(X.T, d_z1) / m  # Shape: (784, 30)
        d_b1 = np.sum(d_z1, axis=0, keepdims=True) / m  # Shape: (1, 30)
        # 그래디언트 클리핑
        clip_value = 1.0
        d_W1 = np.clip(d_W1, -clip_value, clip_value)
        d_b1 = np.clip(d_b1, -clip_value, clip_value)
        d_W2 = np.clip(d_W2, -clip_value, clip_value)
        d_b2 = np.clip(d_b2, -clip_value, clip_value)
        # 가중치와 편향 업데이트
        self.W1 -= learning_rate * d_W1
        self.b1 -= learning_rate * d_b1
        self.W2 -= learning_rate * d_W2
        self.b2 -= learning_rate * d_b2

    def train(self, X_train, Y_train, X_test, y_test, epochs=10000, learning_rate=1.0):
        """모델 학습 및 평가"""
        losses = []
        accuracies = []
        eval_epochs = list(range(0, 1001, 10))  # 0부터 1000까지 10 단위로 평가
        for epoch in range(epochs):
            # 학습 데이터 순전파
            Y_pred_train = self.forward(X_train)
            loss = self.compute_loss(Y_pred_train, Y_train)
            if np.isnan(loss):
                print(f"에포크 {epoch}에서 학습 중단: 손실이 NaN입니다")
                break
            losses.append(loss)

            # 역전파
            self.backward(X_train, Y_train, learning_rate)

            # 테스트 데이터 평가
            if epoch in eval_epochs:
                Y_pred_test = self.forward(X_test)
                eval_loss = self.compute_loss(Y_pred_test, Y_test)
                accuracy = self.compute_accuracy(Y_pred_test, y_test)
                accuracies.append(accuracy)
                print(f"n = {epoch:5d}, Eval = {eval_loss:.4f}, Accurracy = {accuracy:5.1f}%")

        return losses, accuracies, eval_epochs

# MLP 학습
mlp = MLP(input_size=D, hidden_size=L, output_size=M)
losses, accuracies, eval_epochs = mlp.train(X_train, Y_train, X_test, y_test, epochs=10000, learning_rate=1.0)

# 시각화: 학습 손실 및 테스트 정확도
plt.figure(figsize=(12, 5))

# 학습 손실 플롯
plt.subplot(1, 2, 1)
plt.plot(losses)
plt.title("에포크에 따른 학습 손실")
plt.xlabel("에포크")
plt.ylabel("크로스엔트로피 손실")
plt.grid(True)

# 테스트 정확도 플롯
plt.subplot(1, 2, 2)
plt.plot(eval_epochs, accuracies)
plt.title("에포크에 따른 테스트 정확도")
plt.xlabel("에포크")
plt.ylabel("정확도 (%)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
